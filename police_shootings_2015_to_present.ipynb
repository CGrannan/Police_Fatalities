Now let's look at how the data breaks down over time. We will look at total fatalities by month and by year

```python
months = present_data.set_index(present_data.date)
months = months.resample('m', kind='period').count()['id']

fig, ax = plt.subplots()
ax = months.plot(kind='bar', figsize=(20, 8))

#Reduce x-axis clutter by limiting the number of x-ticks:
ticks = ax.get_xticks()
labels = ax.get_xticklabels()
n = len(ticks) // 25  # Show 25 ticks, every third one.
ax.set_xticks(ticks[::3]) 
ax.set_xticklabels(labels[::3])

plt.title('Monthly Number of Shootings')
plt.xlabel('Month')
plt.ylabel('Number of Shootings')
plt.show()
```

```python
years = present_data.set_index(present_data.date)
years = years.resample('y', kind='period').count()['id']
years.plot(kind='bar', figsize=(20, 8))

plt.title('Number of Shootings By Year: 2015-2020')
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.show()
```

That is surprisingly consistent.

Now I will check out how the numbers of shootings change yearly in respect to gender, race and state.

```python
yearly_gender = present_data.groupby([present_data.date.dt.year, 'gender']).id.count()
yearly_gender.unstack().plot(figsize=(12,8))

plt.title('Number of Shootings Yearly By Gender: 2015-2020')
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.show()
```

```python
yearly_race = present_data.groupby([present_data.date.dt.year, 'race']).id.count()
yearly_race.unstack().plot(figsize=(12,8))

plt.title('Number of Shootings Yearly By Race: 2015-2020')
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.show()
```

```python
yearly_state = present_data.groupby([present_data.date.dt.year, 'state']).id.count()
yearly_state.unstack().plot(figsize=(20,10))

plt.title('Number of Shootings Yearly By State: 2015-2020')
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.show()
```

Now I will be looking at the same graphs for the older shooting data (2000-2016).

```python
older_data.State.value_counts('normalize').plot(kind='bar', figsize=(20,10))
plt.title('Number of Fatalities by State')
plt.xlabel('State')
plt.ylabel('Number of Fatalities')
plt.show()
```

```python
older_months = older_data.set_index(older_data.Date)
older_months = older_months.resample('m', kind='period').count()['UID']

fig, ax = plt.subplots()
ax = older_months.plot(kind='bar', figsize=(20, 12))

# Again reduce x-ticks to make graph legible
ticks = ax.get_xticks()
labels = ax.get_xticklabels()
n = len(ticks) // 16  # Show 16 ticks, every 12th one.
ax.set_xticks(ticks[6::12])
ax.set_xticklabels(labels[6::12])

plt.title('Number of Shootings by Month: 2000-2016')
plt.xlabel('Month')
plt.ylabel('Number of Shootings')
plt.show()
```
